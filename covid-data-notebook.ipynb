```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data from URL (or local CSV)
url = "/Your/Data/Path/owid-covid-data.csv"
df = pd.read_csv(url)

#  Set your desired country
country_name = "United States"  # Change this to any country

# Filter the DataFrame for that country
country_df = df[df["location"] == country_name]

# Print the first 100 rows
print("First 100 rows for United States")
print(country_df.head(100))

# Convert 'date' column to datetime
country_df = df[df['location'] == 'India'].copy()
country_df['date'] = pd.to_datetime(country_df['date'])


# Set up the plot
plt.figure(figsize=(14, 6))  # Widen the plot

# Plot daily new cases
plt.plot(country_df['date'], country_df['new_cases'], label='New Cases', color='blue')

# Format the x-axis date labels
plt.xticks(rotation=45)  # Tilt the labels
plt.locator_params(axis='x', nbins=15)  # Show fewer ticks (adjust 1020 as needed)

# Add title and labels
plt.title("Daily New COVID-19 Cases - United States")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
